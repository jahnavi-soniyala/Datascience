```python
import pandas as pd
```

```python
df=pd.read_csv("dataset.csv")
df
```

```python
df.dtypes
```

```python
#votes, Average Cost for two, Price range, Rating text 
df['Rating text'].values
```

```python
rating=pd.get_dummies(df['Rating text'],prefix='Rating_text',dtype=int)
df=pd.concat([df,rating],axis=1)
df.drop(columns=['Rating text'],inplace=True)
df
```

```python
city=pd.get_dummies(df['City'],dtype=int)
df= pd.concat([df,city],axis=1)
df.drop(columns=['City'],inplace=True)
df
```

```python
cuisine= pd.get_dummies(df['Cuisines'],dtype=int)
df= pd.concat([df,cuisine],axis=1)
df
```

```python
#input = votes, Average Cost for two, Price range, Rating text, Country code, City, Cuisine, Has Table booking, Has Online delivery,,  'Is delivering now','Switch to order menu',
input=df.drop(columns=['Restaurant ID', 'Restaurant Name','Address','Locality', 'Locality Verbose', 'Longitude', 'Latitude', 'Cuisines', 'Currency', 'Aggregate rating', 'Rating color' ])
```

```python
output=df["Aggregate rating"]
```

```python
from sklearn.preprocessing import LabelEncoder
le_Tablebooking = LabelEncoder()
input['Has Table booking new']=le_Tablebooking.fit_transform(input['Has Table booking'])
le_onlinedelivery = LabelEncoder()
input['Has Online delivery new']=le_onlinedelivery.fit_transform(input['Has Online delivery'])
le_deliverynow = LabelEncoder()
input['Is delivering now new']=le_deliverynow.fit_transform(input['Is delivering now'])
le_menuswitch = LabelEncoder()
input['Switch to order menu new']=le_deliverynow.fit_transform(input['Switch to order menu'])
input.drop(columns=['Has Table booking','Has Online delivery','Is delivering now','Switch to order menu'],inplace=True)
```

```python
input
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(input,output,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_s = scaler.fit_transform(xtrain)
x_test_s = scaler.transform(xtest)
```

```python
from sklearn.tree import DecisionTreeRegressor
model_dr=DecisionTreeRegressor()
model_dr.fit(x_train_s,ytrain)
```

```python
model_dr.score(x_test_s,ytest)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_rdf=RandomForestRegressor()
model_rdf.fit(x_train_s,ytrain)
```

```python
model_rdf.score(x_test_s,ytest)
```

```python
x_scaled = scaler.fit_transform(input)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
models = {
    'Decision Tree': model_dr,
    'Linear Regression': model,
    'Random Forest': model_rdf
}
```

```python
for name, model in models.items():
    scores = cross_val_score(model, x_scaled, output, cv=5, scoring='r2')  
    print(name,":",model,'is the best model with mean :' ,scores.mean(),"ans std:", scores.std())
```
